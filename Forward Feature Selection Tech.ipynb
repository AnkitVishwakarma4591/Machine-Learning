```python
import pandas as pd
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df.head(3)
```

```python
x = df.iloc[:,:-1]
y = df['Outcome']
```

```python
x.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
fs = SequentialFeatureSelector(lr,k_features=6,forward=True)
# if we have to use backward elimination then Forward = False
```

```python
fs.fit(x,y)
```

```python
fs.feature_names
```

```python
fs.k_feature_names_
```

```python
fs.k_score_
```
