```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('polynomial.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
x = df.iloc[:,:-1]
y = df['Ice Cream Sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor(criterion='absolute_error',max_depth=6,splitter='best')
```

```python
dt.fit(x_train, y_train)
```

```python
dt.score(x_test, y_test)*100
```

```python
dt.score(x_train, y_train)*100
```

## Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
dataset = {
    "criterion" : ["squared_error", "friedman_mse", "absolute_error"],
    "splitter" : ["best", "random"],
    "max_depth" : [i for i in range(2,20)]
}
```

## Randomized Search CV

```python
rd = RandomizedSearchCV(DecisionTreeRegressor(),param_distributions=dataset,n_iter=20)
```

```python
rd.fit(x_train, y_train)
```

```python
rd.best_params_
```

```python
rd.best_score_
```

## Gridsearch CV

```python
gd = GridSearchCV(DecisionTreeRegressor(), param_grid= dataset)
```

```python
gd.fit(x_train, y_train)
```

```python
gd.best_params_
```

```python
gd.best_score_
```
