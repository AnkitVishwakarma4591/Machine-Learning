```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('loan.csv')
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
sns.distplot(df['ApplicantIncome'])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler()
```

```python
ms.fit(df[['ApplicantIncome']])
```

```python
ms.transform(df[['ApplicantIncome']])
```

```python
df['ApplicantIncome_ms'] = pd.DataFrame(ms.transform(df[['ApplicantIncome']]),columns=['X'])
```

```python
df.head(3)
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
sns.distplot(df['ApplicantIncome'])
plt.title("Before")

plt.subplot(1,2,2)
sns.distplot(df['ApplicantIncome_ms'])
plt.title("After")

plt.show()
```
