```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv('Naive Bayes.csv')
```

```python
df.head()
```

```python
# Checking Distribution of data
```

```python
sns.kdeplot(data=df['CGPA'])
plt.show()
```

```python
sns.kdeplot(data=df['Score'])
plt.show()
```

```python
plt.figure(figsize=[7,5])
sns.scatterplot(x=df['CGPA'],y=df['Score'],hue=df['Placed'])
plt.show()
```

```python
x = df.iloc[:,:-1]
y = df['Placed']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

## Gaussian Naive Bayes

```python
gnb = GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100, gnb.score(x_train,y_train)*100
```

```python
plt.figure(figsize=[7,5])
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
plt.show()
```

```python
gnb.predict([[7.53,7.97]]) # 7.53	7.97	1
```

## Multinomial Naive Bayes

```python
mnb = MultinomialNB()
```

```python
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_test,y_test)*100 , mnb.score(x_train,y_train)*100
```

```python
plt.figure(figsize=[7,5])
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mnb)
plt.show()
```

## Bernoulli Naive Bayes

```python
bnb = BernoulliNB()
```

```python
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_test,y_test)*100 , bnb.score(x_train,y_train)*100
```

```python
plt.figure(figsize=[7,5])
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb)
plt.show()
```
