```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
sc.fit(df[['ApplicantIncome']])
```

```python
sc.transform(df[['ApplicantIncome']])
```

```python
df['ApplicantIncome_ss'] = pd.DataFrame(sc.transform(df[['ApplicantIncome']]),columns=['X'])
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
sns.distplot(df['ApplicantIncome'])
plt.title("Brfore")

plt.subplot(1,2,2)
sns.distplot(df['ApplicantIncome_ss'])
plt.title("After")

plt.show()
```
